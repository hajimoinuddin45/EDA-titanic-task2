```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
# Load the Titanic dataset
df = pd.read_csv("C:/Users/Haji Moinuddin/OneDrive/Desktop/Titanic-Dataset.csv")
print("Dataset loaded successfully!")

# Display basic statistics
print("Summary statistics:\n", df.describe(include='all'))
```

```python
# Check for missing values
print("\nMissing values:\n", df.isnull().sum())
```

```python
# Histogram for Age and Fare
plt.figure(figsize=(14, 5))
plt.subplot(1, 2, 1)
sns.histplot(df['Age'].dropna(), kde=True)
plt.title('Age Distribution')



plt.subplot(1, 2, 2)
sns.histplot(df['Fare'], kde=True)
plt.title('Fare Distribution')

plt.tight_layout()
plt.show()
```

```python
# Boxplots to find outliers
plt.figure(figsize=(14, 5))
plt.subplot(1, 2, 1)
sns.boxplot(x=df['Age'])
plt.title('Boxplot - Age')

plt.subplot(1, 2, 2)
sns.boxplot(x=df['Fare'])
plt.title('Boxplot - Fare')

plt.tight_layout()
plt.show()
```

```python
# Correlation matrix
corr_matrix = df.corr(numeric_only=True)
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
# Pairplot for selected features
sns.pairplot(df[['Age', 'Fare', 'Pclass', 'Survived']].dropna(), hue='Survived')
plt.suptitle("Pairplot of Key Features", y=1.02)
plt.show()
```

```python
# Survival count by gender (basic inference from visuals)
sns.countplot(x='Sex', hue='Survived', data=df)
plt.title('Survival by Gender')
plt.show()
```

```python
# Survival count by class
sns.countplot(x='Pclass', hue='Survived', data=df)
plt.title('Survival by Passenger Class')
plt.show()
```
